```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from binance_data_loader import BinanceDataLoader
from tqdm import tqdm
```

```python
data_loader = BinanceDataLoader(
    data_directory=r"C:\Users\USER\Documents\Binance_related\dailytickerdata2020",
    min_records=30,
    min_volume=1e5,
    start_date="2021-01-01",
    end_date=None
)
```

```python
price = data_loader.get_price_matrix()
price
```

```python
n =30
rets = price.pct_change(n)
equity = [1.0]
rets
```

```python
# Get volume data for universe selection
volume_data = {}
for ticker in data_loader.get_universe():
    ticker_data = data_loader._crypto_universe[ticker]['data']
    volume_data[ticker] = ticker_data['volume'].reindex(price.index)

volume_matrix = pd.DataFrame(volume_data, index=price.index)
# Calculate BTC 90-day total return for market filter
btc_90d_return = price["BTCUSDT"].pct_change(90)

k = 10  # Number of coins to go long/short
n_universe = 50  # Top N volume universe to select from each day

# Calculate rolling volatility for weighting (using 30-day window)
vol_window = 30
volatility = price.pct_change().rolling(vol_window).std()

for i in tqdm(range(len(rets)-1)):
    current_date = price.index[i]
    # BTC Market Filter: Skip trading if BTC 90-day return is negative
    if not pd.isna(btc_90d_return.iloc[i]) and btc_90d_return.iloc[i] > 0:
        equity.append(equity[-1])  # Hold cash during crypto bull markets
        continue
    # Get valid volume data for current date
    current_volumes = volume_matrix.iloc[i]
    valid_volumes = current_volumes.dropna()
    
    # Skip if no volume data available
    if len(valid_volumes) == 0:
        equity.append(equity[-1])
        continue
    
    # Select top N_universe by volume for this day
    top_volume_tickers = valid_volumes.nlargest(n_universe).index
    
    # Get returns for the volume-filtered universe
    row = rets.iloc[i]
    filtered_returns = row[top_volume_tickers].dropna()
    
    # Skip if no valid returns in the volume universe
    if len(filtered_returns) == 0:
        equity.append(equity[-1])
        continue

    # Mean reversion: long worst performers, short best performers
    long_coins = filtered_returns.nsmallest(k).index   # worst performers (for mean reversion)
    short_coins = filtered_returns.nlargest(k).index  # best performers (for mean reversion)
   
        # Calculate inverse volatility weights
    long_vols = volatility.iloc[i][long_coins]
    short_vols = volatility.iloc[i][short_coins]

    ''' # keep only coins with valid volatility (so we don't trade names with missing data)
    long_valid_idx = long_vols.dropna().index
    short_valid_idx = short_vols.dropna().index

    if len(long_valid_idx) == 0 or len(short_valid_idx) == 0:
        equity.append(equity[-1])
        continue

    # equal weight per coin; long side sums to 0.5 and short side sums to 0.5
    long_n = len(long_valid_idx)
    short_n = len(short_valid_idx)

    long_weights = pd.Series(0.5 / long_n, index=long_valid_idx)
    short_weights = pd.Series(0.5 / short_n, index=short_valid_idx)'''
   # Remove coins with NaN volatility
    long_valid = long_vols.dropna()
    short_valid = short_vols.dropna()
    
    if len(long_valid) == 0 or len(short_valid) == 0:
        equity.append(equity[-1])
        continue
    
    # Inverse volatility weights (higher weight for lower vol)
    long_weights = 1 / long_valid
    short_weights = 1 / short_valid
    
    # Normalize so each side sums to 0.5 (equal long/short exposure)
    long_weights =  long_weights/long_weights.sum() * 0.5
    short_weights =  short_weights/short_weights.sum() * 0.5

    # Calculate weighted returns for the next day
    long_return = 0
    short_return = 0
    
    for coin in long_weights.index:
        p0 = price[coin].iat[i]
        p1 = price[coin].iat[i+1]
        if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
            coin_return = (p1 - p0) / p0
            long_return += long_weights[coin] * coin_return

    for coin in short_weights.index:
        p0 = price[coin].iat[i]
        p1 = price[coin].iat[i+1]
        if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
            coin_return = (p1 - p0) / p0
            short_return += short_weights[coin] * coin_return

    # Portfolio return: long - short - fees
    portfolio_return = long_return - short_return - 0.0005  # 5bps fees
    equity.append(equity[-1] * (1 + portfolio_return))

#plt.plot(equity)
plt.plot(price.index[:len(equity)], equity)
plt.xlabel("Date")
plt.ylabel("Equity")
plt.title("Strategy Equity Curve")
plt.xticks(rotation=45)
plt.show()
```

```python
# Parameter optimization loop with equity curve plotting
import itertools

# Parameter ranges to test
n_values = [10, 15, 20, 30,]  # Lookback periods
k_values = [5, 8, 10]        # Number of positions per side
n_universe = 50  # Keep volume universe fixed

results = []
equity_curves = {}  # Store equity curves for plotting

# Get volume data once (reused for all parameter combinations)
volume_data = {}
for ticker in data_loader.get_universe():
    ticker_data = data_loader._crypto_universe[ticker]['data']
    volume_data[ticker] = ticker_data['volume'].reindex(price.index)

volume_matrix = pd.DataFrame(volume_data, index=price.index)

print("Testing parameter combinations...")
for n, k in tqdm(list(itertools.product(n_values, k_values))):
    # Calculate returns for this lookback period
    rets = price.pct_change(n)
    equity = [1.0]
    
    for i in range(len(rets)-1):
        current_date = price.index[i]
        
        # Get valid volume data for current date
        current_volumes = volume_matrix.iloc[i]
        valid_volumes = current_volumes.dropna()
        
        # Skip if no volume data available
        if len(valid_volumes) == 0:
            equity.append(equity[-1])
            continue
        
        # Select top N_universe by volume for this day
        top_volume_tickers = valid_volumes.nlargest(n_universe).index
        
        # Get returns for the volume-filtered universe
        row = rets.iloc[i]
        filtered_returns = row[top_volume_tickers].dropna()
        
        # Skip if no valid returns in the volume universe
        if len(filtered_returns) == 0:
            equity.append(equity[-1])
            continue

        # Mean reversion: long worst performers, short best performers
        long_coins = filtered_returns.nsmallest(k).index   # worst performers
        short_coins = filtered_returns.nlargest(k).index  # best performers
       
        # Calculate returns for the next day
        long_returns = []
        short_returns = []

        for coin in long_coins:
            p0 = price[coin].iat[i]
            p1 = price[coin].iat[i+1]
            if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
                long_returns.append((p1 - p0)/p0)

        for coin in short_coins:
            p0 = price[coin].iat[i]
            p1 = price[coin].iat[i+1]
            if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
                short_returns.append((p1 - p0)/p0)

        # If no valid trades, hold flat
        if not long_returns or not short_returns:
            equity.append(equity[-1])
            continue

        ret_next_long = np.mean(long_returns)
        ret_next_short = np.mean(short_returns)
        equity.append(equity[-1] * (1 + 0.5*ret_next_long - 0.5*ret_next_short - 0.0005))
    
    # Store equity curve
    equity_series = pd.Series(equity, index=price.index[:len(equity)])
    equity_curves[f"n={n}, k={k}"] = equity_series
    
    # Calculate performance metrics
    total_return = equity_series.iloc[-1] - 1
    daily_returns = equity_series.pct_change().dropna()
    volatility = daily_returns.std() * np.sqrt(252)  # Annualized
    sharpe = daily_returns.mean() / daily_returns.std() * np.sqrt(252) if daily_returns.std() > 0 else 0
    max_dd = (equity_series / equity_series.cummax() - 1).min()
    
    results.append({
        'n': n,
        'k': k, 
        'total_return': total_return,
        'volatility': volatility,
        'sharpe': sharpe,
        'max_drawdown': max_dd,
        'final_equity': equity_series.iloc[-1]
    })

# Convert to DataFrame and display best results
results_df = pd.DataFrame(results)
print("\nTop 10 parameter combinations by Sharpe ratio:")
top_sharpe = results_df.nlargest(10, 'sharpe')[['n', 'k', 'total_return', 'sharpe', 'max_drawdown']]
print(top_sharpe)

print("\nTop 10 parameter combinations by total return:")
top_return = results_df.nlargest(10, 'total_return')[['n', 'k', 'total_return', 'sharpe', 'max_drawdown']]
print(top_return)

# Plot equity curves for top performers
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))

# Plot top 5 by Sharpe ratio
print("\nPlotting top 5 strategies by Sharpe ratio...")
top_5_sharpe = results_df.nlargest(5, 'sharpe')
for _, row in top_5_sharpe.iterrows():
    label = f"n={row['n']}, k={row['k']} (Sharpe: {row['sharpe']:.2f})"
    equity_curves[f"n={row['n']}, k={row['k']}"].plot(ax=ax1, label=label, alpha=0.8)

ax1.set_title("Top 5 Strategies by Sharpe Ratio")
ax1.set_ylabel("Equity")
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot top 5 by total return
print("Plotting top 5 strategies by total return...")
top_5_return = results_df.nlargest(5, 'total_return')
for _, row in top_5_return.iterrows():
    label = f"n={row['n']}, k={row['k']} (Return: {row['total_return']:.1%})"
    equity_curves[f"n={row['n']}, k={row['k']}"].plot(ax=ax2, label=label, alpha=0.8)

ax2.set_title("Top 5 Strategies by Total Return")
ax2.set_ylabel("Equity")
ax2.set_xlabel("Date")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Add BTC comparison
btc_cumret = (1 + price["BTCUSDT"].pct_change().fillna(0)).cumprod()
print(f"\nBTC Buy & Hold Return: {(btc_cumret.iloc[-1] - 1):.1%}")

# Plot best strategy vs BTC
best_strategy = results_df.loc[results_df['sharpe'].idxmax()]
best_label = f"Best Strategy: n={best_strategy['n']}, k={best_strategy['k']}"

plt.figure(figsize=(12, 6))
equity_curves[f"n={best_strategy['n']}, k={best_strategy['k']}"].plot(label=best_label, linewidth=2)
btc_cumret.plot(label="BTC Buy & Hold", linewidth=2)
plt.title("Best Mean Reversion Strategy vs BTC")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
