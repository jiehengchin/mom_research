```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from binance_data_loader import BinanceDataLoader
from tqdm import tqdm
```

```python
data_loader = BinanceDataLoader(
    data_directory=r"C:\Users\USER\Documents\Binance_related\dailytickerdata2020",
    min_records=30,
    min_volume=1e5,
    start_date="2021-01-01",
    end_date=None
)
```

```python
price = data_loader.get_price_matrix()
price
```

```python
n =14
rets = price.shift(1).pct_change(n)
equity = [1.0]
rets
```

```python
# Get volume data for universe selection with 20-day rolling average
volume_data = {}
for ticker in data_loader.get_universe():
    ticker_data = data_loader._crypto_universe[ticker]['data']
    volume_data[ticker] = ticker_data['volume'].reindex(price.index)

volume_matrix = pd.DataFrame(volume_data, index=price.index)

# Calculate 20-day rolling average volume
rolling_volume_matrix = volume_matrix.rolling(window=20, min_periods=10).mean()

# Calculate rolling volatility for weighting (using 30-day window)
vol_window = 20
volatility = price.pct_change().rolling(vol_window).std()

# Calculate BTC 90-day total return for market filter
btc_90d_return = price["BTCUSDT"].pct_change(90)

k = 10  # Number of coins to go long/short
n_universe = 55  # Top N volume universe to select from each day

# Main backtest loop - using mean-variance optimizer with regularization (fallback to score/vol)
for i in range(n, len(price)-1):
    current_date = price.index[i]

    # BTC Market Filter: Skip trading if BTC 90-day return is negative
    if not pd.isna(btc_90d_return.iloc[i]) and btc_90d_return.iloc[i] < 0:
        equity.append(equity[-1])  # Hold cash during crypto bear markets
        continue

    # Get 20-day rolling average volume for current date
    current_volumes = rolling_volume_matrix.iloc[i].dropna()

    # Skip if no volume data available
    if len(current_volumes) == 0:
        equity.append(equity[-1])
        continue

    # Select top N_universe by 20-day rolling volume for this day
    top_volume_tickers = current_volumes.nlargest(n_universe).index

    # Get momentum returns for the volume-filtered universe (score)
    row = rets.iloc[i]
    filtered_returns = row[top_volume_tickers].dropna()

    # Skip if no valid returns in the volume universe
    if len(filtered_returns) == 0:
        equity.append(equity[-1])
        continue

    # Simple momentum ranking: top performers go long, bottom performers go short
    long_coins = filtered_returns.nlargest(k).index   # top k momentum
    short_coins = filtered_returns.nsmallest(k).index # bottom k momentum

    # Build scores for selected coins (momentum returns as score)
    long_scores = filtered_returns.reindex(long_coins)
    short_scores = filtered_returns.reindex(short_coins)

    # Mean-variance optimizer parameters
    mv_lookback = 20  # lookback (days) for estimating mu and Sigma
    reg = 1e-5        # regularization added to diagonal of Sigma
    lambda_param = 2.0  # risk aversion / scaling for Sigma

    # Use the top-volume universe for covariance estimation
    chosen_universe = top_volume_tickers

    # Build returns dataframe ending at i (shifted to avoid lookahead)
    start_idx = max(0, i - mv_lookback + 1)
    returns_df = price[chosen_universe].pct_change().shift(1).iloc[start_idx:i+1].dropna(how='all')

    # If insufficient history to estimate cov, fall back to score/vol heuristic
    if returns_df.shape[0] < 10:
        long_vols = price[long_coins].pct_change().rolling(vol_window).std().iloc[i]
        short_vols = price[short_coins].pct_change().rolling(vol_window).std().iloc[i]
        long_ratio = (long_scores / long_vols).replace([np.inf, -np.inf], np.nan).dropna()
        short_ratio = ((-short_scores) / short_vols).replace([np.inf, -np.inf], np.nan).dropna()
        if len(long_ratio) == 0 or len(short_ratio) == 0:
            equity.append(equity[-1])
            continue
        long_weights = long_ratio / long_ratio.sum() * 0.5
        short_weights = short_ratio / short_ratio.sum() * 0.5
    else:
        # Estimate mu and Sigma from recent returns
        mu = returns_df.mean()
        Sigma = returns_df.cov().values + reg * np.eye(len(returns_df.columns))

        try:
            raw_w = np.linalg.solve(lambda_param * Sigma, mu.values)
        except Exception:
            # fallback if Sigma is singular or solve fails
            long_vols = price[long_coins].pct_change().rolling(vol_window).std().iloc[i]
            short_vols = price[short_coins].pct_change().rolling(vol_window).std().iloc[i]
            long_ratio = (long_scores / long_vols).replace([np.inf, -np.inf], np.nan).dropna()
            short_ratio = ((-short_scores) / short_vols).replace([np.inf, -np.inf], np.nan).dropna()
            if len(long_ratio) == 0 or len(short_ratio) == 0:
                equity.append(equity[-1])
                continue
            long_weights = long_ratio / long_ratio.sum() * 0.5
            short_weights = short_ratio / short_ratio.sum() * 0.5
        else:
            raw_w_series = pd.Series(raw_w, index=mu.index)

            # Split positive (long) and negative (short) raw weights
            longs_raw = raw_w_series[raw_w_series > 0]
            shorts_raw = raw_w_series[raw_w_series < 0]

            # If either side empty, skip the day
            if len(longs_raw) == 0 or len(shorts_raw) == 0:
                equity.append(equity[-1])
                continue

            # Convert to positive magnitudes for allocation and normalize each side to 0.5
            long_weights = longs_raw / longs_raw.sum() * 0.5
            short_weights = (-shorts_raw) / (-shorts_raw).sum() * 0.5

    # Calculate weighted returns for the next day
    long_return = 0
    short_return = 0

    for coin in long_weights.index:
        p0 = price[coin].iat[i]
        p1 = price[coin].iat[i+1]
        if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
            coin_return = (p1 - p0) / p0
            long_return += long_weights[coin] * coin_return

    for coin in short_weights.index:
        p0 = price[coin].iat[i]
        p1 = price[coin].iat[i+1]
        if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
            coin_return = (p1 - p0) / p0
            short_return += short_weights[coin] * coin_return

    # Portfolio return: long - short - fees
    portfolio_return = long_return - short_return - 0.0005  # 5bps fees
    equity.append(equity[-1] * (1 + portfolio_return))
```

```python
plt.plot(price.index[:len(equity)], equity)
plt.xlabel("Date")
plt.ylabel("Equity")
plt.title("Strategy Equity Curve")
plt.xticks(rotation=45)
plt.show()
```

```python
# Diagnostic: compare mean-variance weights vs score/vol heuristic for the next N valid signal days
from collections import defaultdict

def compute_mv_weights_for_day(i, mv_lookback=60, reg=1e-5, lambda_param=1.0):
    current_volumes = rolling_volume_matrix.iloc[i].dropna()
    if len(current_volumes) == 0:
        return None
    top_volume_tickers = current_volumes.nlargest(n_universe).index
    row = rets.iloc[i]
    filtered_returns = row[top_volume_tickers].dropna()
    if len(filtered_returns) == 0:
        return None

    # estimate mu/Sigma
    start_idx = max(0, i - mv_lookback + 1)
    returns_df = price[top_volume_tickers].pct_change().shift(1).iloc[start_idx:i+1].dropna(how='all')
    if returns_df.shape[0] < 10:
        return None
    mu = returns_df.mean()
    Sigma = returns_df.cov().values + reg * np.eye(len(returns_df.columns))
    try:
        raw_w = np.linalg.solve(lambda_param * Sigma, mu.values)
    except Exception:
        return None
    raw_w_series = pd.Series(raw_w, index=mu.index)
    longs_raw = raw_w_series[raw_w_series > 0]
    shorts_raw = raw_w_series[raw_w_series < 0]
    if len(longs_raw) == 0 or len(shorts_raw) == 0:
        return None
    long_w = longs_raw / longs_raw.sum() * 0.5
    short_w = (-shorts_raw) / (-shorts_raw).sum() * 0.5
    return long_w, short_w


def compute_scorevol_weights_for_day(i):
    current_volumes = rolling_volume_matrix.iloc[i].dropna()
    if len(current_volumes) == 0:
        return None
    top_volume_tickers = current_volumes.nlargest(n_universe).index
    row = rets.iloc[i]
    filtered_returns = row[top_volume_tickers].dropna()
    if len(filtered_returns) == 0:
        return None
    long_coins = filtered_returns.nlargest(k).index
    short_coins = filtered_returns.nsmallest(k).index
    long_vols = volatility.iloc[i][long_coins]
    short_vols = volatility.iloc[i][short_coins]
    long_scores = filtered_returns.reindex(long_coins)
    short_scores = filtered_returns.reindex(short_coins)
    long_ratio = (long_scores / long_vols).replace([np.inf, -np.inf], np.nan).dropna()
    short_ratio = ((-short_scores) / short_vols).replace([np.inf, -np.inf], np.nan).dropna()
    if len(long_ratio) == 0 or len(short_ratio) == 0:
        return None
    long_w = long_ratio / long_ratio.sum() * 0.5
    short_w = short_ratio / short_ratio.sum() * 0.5
    return long_w, short_w

# Run diagnostics for next N valid signal days
N = 10
printed = 0
mv_perf = []
sv_perf = []

diagnostics = []
for i in range(n, len(price)-1):
    mv = compute_mv_weights_for_day(i)
    sv = compute_scorevol_weights_for_day(i)
    if mv is None and sv is None:
        continue
    diag = {'date': price.index[i].date()}
    if mv is not None:
        long_mv, short_mv = mv
        diag['mv_long_sum'] = long_mv.sum()
        diag['mv_short_sum'] = short_mv.sum()
        diag['mv_long'] = long_mv.to_dict()
        diag['mv_short'] = short_mv.to_dict()
        # compute next-day portfolio return for MV
        lr = sum((price.loc[price.index[i+1], c] - price.loc[price.index[i], c]) / price.loc[price.index[i], c] * w for c, w in long_mv.items() if c in price.columns)
        sr = sum((price.loc[price.index[i+1], c] - price.loc[price.index[i], c]) / price.loc[price.index[i], c] * w for c, w in short_mv.items() if c in price.columns)
        mv_ret = lr - sr - 0.0005
        mv_perf.append(mv_ret)
        diag['mv_ret'] = mv_ret
    if sv is not None:
        long_sv, short_sv = sv
        diag['sv_long_sum'] = long_sv.sum()
        diag['sv_short_sum'] = short_sv.sum()
        diag['sv_long'] = long_sv.to_dict()
        diag['sv_short'] = short_sv.to_dict()
        lr = sum((price.loc[price.index[i+1], c] - price.loc[price.index[i], c]) / price.loc[price.index[i], c] * w for c, w in long_sv.items() if c in price.columns)
        sr = sum((price.loc[price.index[i+1], c] - price.loc[price.index[i], c]) / price.loc[price.index[i], c] * w for c, w in short_sv.items() if c in price.columns)
        sv_ret = lr - sr - 0.0005
        sv_perf.append(sv_ret)
        diag['sv_ret'] = sv_ret

    diagnostics.append(diag)
    printed += 1
    if printed >= N:
        break

# Print diagnostics
for d in diagnostics:
    print('\nDate:', d['date'])
    if 'mv_long_sum' in d:
        print(' MV long sum=', d['mv_long_sum'], ' MV short sum=', d['mv_short_sum'], ' MV ret=', d.get('mv_ret'))
        print('  MV longs:', d['mv_long'])
        print('  MV shorts:', d['mv_short'])
    if 'sv_long_sum' in d:
        print(' SV long sum=', d['sv_long_sum'], ' SV short sum=', d['sv_short_sum'], ' SV ret=', d.get('sv_ret'))
        print('  SV longs:', d['sv_long'])
        print('  SV shorts:', d['sv_short'])

# Small performance summary
if len(mv_perf) > 0:
    print('\nMV avg return over sample:', np.mean(mv_perf), ' std:', np.std(mv_perf))
if len(sv_perf) > 0:
    print('SV avg return over sample:', np.mean(sv_perf), ' std:', np.std(sv_perf))

print('\nDiagnostics complete')
```
