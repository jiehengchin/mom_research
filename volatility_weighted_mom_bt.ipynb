```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from binance_data_loader import BinanceDataLoader
from tqdm import tqdm
```

```python
data_loader = BinanceDataLoader(
    data_directory=r"C:\Users\USER\Documents\Binance_related\dailytickerdata2020",
    min_records=30,
    min_volume=1e5,
    start_date="2022-01-01",
    end_date=None
)
```

```python
price = data_loader.get_price_matrix()
price
```

```python
n =14
rets = price.shift(1).pct_change(n)
equity = [1.0]
rets
```

```python
# Get volume data for universe selection with 20-day rolling average
volume_data = {}
for ticker in data_loader.get_universe():
    ticker_data = data_loader._crypto_universe[ticker]['data']
    volume_data[ticker] = ticker_data['volume'].reindex(price.index)

volume_matrix = pd.DataFrame(volume_data, index=price.index)

# Calculate 20-day rolling average volume
rolling_volume_matrix = volume_matrix.rolling(window=20, min_periods=10).mean()

# Calculate rolling volatility for weighting (using 30-day window)
vol_window = 30
volatility = price.pct_change().rolling(vol_window).std()

# Calculate BTC 90-day total return for market filter
btc_90d_return = price["BTCUSDT"].pct_change(90)

# Calculate rolling beta with BTC as benchmark (using 30-day window)
beta_window = 20
btc_returns = price["BTCUSDT"].pct_change()
portfolio_betas = {}

k = 10  # Number of coins to go long/short
n_universe = 50  # Top N volume universe to select from each day

# Track current positions and last rebalance date
current_long_weights = pd.Series(dtype=float)
current_short_weights = pd.Series(dtype=float)
last_rebalance_date = None

# Main backtest loop - using volatility-based position sizing
for i in range(n, len(price)-1):
    current_date = price.index[i]

    # BTC Market Filter: Skip trading if BTC 90-day return is negative
    if not pd.isna(btc_90d_return.iloc[i]) and btc_90d_return.iloc[i] < 0:
        equity.append(equity[-1])  # Hold cash during crypto bear markets
        continue

    # Get 20-day rolling average volume for current date
    current_volumes = rolling_volume_matrix.iloc[i].dropna()

    # Skip if no volume data available
    if len(current_volumes) == 0:
        equity.append(equity[-1])
        continue

    # Select top N_universe by 20-day rolling volume for this day
    top_volume_tickers = current_volumes.nlargest(n_universe).index

    # Get momentum returns for the volume-filtered universe (score)
    row = rets.iloc[i]
    filtered_returns = row[top_volume_tickers].dropna()

    # Skip if no valid returns in the volume universe
    if len(filtered_returns) == 0:
        equity.append(equity[-1])
        continue

    # Calculate portfolio beta with BTC if we have current positions
    should_rebalance = True
    if len(current_long_weights) > 0 or len(current_short_weights) > 0:
        # Calculate portfolio returns for beta calculation
        start_beta_idx = max(0, i - beta_window + 1)
        
        portfolio_returns = []
        for j in range(start_beta_idx, i + 1):
            if j == 0:
                portfolio_returns.append(0)
                continue
                
            portfolio_return = 0
            
            # Calculate long portfolio return
            for coin, weight in current_long_weights.items():
                if coin in price.columns:
                    p_prev = price[coin].iloc[j-1]
                    p_curr = price[coin].iloc[j]
                    if np.isfinite(p_prev) and p_prev != 0 and np.isfinite(p_curr):
                        coin_return = (p_curr - p_prev) / p_prev
                        portfolio_return += weight * coin_return
            
            # Calculate short portfolio return
            for coin, weight in current_short_weights.items():
                if coin in price.columns:
                    p_prev = price[coin].iloc[j-1]
                    p_curr = price[coin].iloc[j]
                    if np.isfinite(p_prev) and p_prev != 0 and np.isfinite(p_curr):
                        coin_return = (p_curr - p_prev) / p_prev
                        portfolio_return += weight * coin_return  # Short positions have negative weight
            
            portfolio_returns.append(portfolio_return)
        
        portfolio_returns = pd.Series(portfolio_returns, index=price.index[start_beta_idx:i+1])
        btc_returns_subset = btc_returns.iloc[start_beta_idx:i+1]
        
        # Calculate beta if we have enough data
        if len(portfolio_returns) >= 10:
            valid_data = pd.DataFrame({'portfolio': portfolio_returns, 'btc': btc_returns_subset}).dropna()
            if len(valid_data) >= 10:
                portfolio_beta = np.cov(valid_data['portfolio'], valid_data['btc'])[0,1] / np.var(valid_data['btc'])
                
                # Only rebalance if beta >= 0.1
                should_rebalance = abs(portfolio_beta) >= 0.0001

    # If not rebalancing, calculate return with current positions
    if not should_rebalance:
        portfolio_return = 0
        
        # Calculate long return
        for coin, weight in current_long_weights.items():
            if coin in price.columns:
                p0 = price[coin].iloc[i]
                p1 = price[coin].iloc[i+1]
                if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
                    coin_return = (p1 - p0) / p0
                    portfolio_return += weight * coin_return
        
        # Calculate short return
        for coin, weight in current_short_weights.items():
            if coin in price.columns:
                p0 = price[coin].iloc[i]
                p1 = price[coin].iloc[i+1]
                if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
                    coin_return = (p1 - p0) / p0
                    portfolio_return += weight * coin_return  # Short weights are negative
        
        # Apply fees only if we had positions
        if len(current_long_weights) > 0 or len(current_short_weights) > 0:
            portfolio_return -= 0.0005  # 5bps fees
        
        equity.append(equity[-1] * (1 + portfolio_return))
        continue

    # Rebalancing logic - select new positions
    long_coins = filtered_returns.nlargest(k).index   # top k momentum
    short_coins = filtered_returns.nsmallest(k).index # bottom k momentum

    # Get volatilities for selected coins
    long_vols = volatility.iloc[i][long_coins].dropna()
    short_vols = volatility.iloc[i][short_coins].dropna()

    # Skip if we don't have volatility data for any coins
    if len(long_vols) == 0 or len(short_vols) == 0:
        equity.append(equity[-1])
        continue

    # Calculate inverse volatility weights
    long_inv_vol = 1.0 / long_vols
    short_inv_vol = 1.0 / short_vols

    # Remove infinite values
    long_inv_vol = long_inv_vol.replace([np.inf, -np.inf], np.nan).dropna()
    short_inv_vol = short_inv_vol.replace([np.inf, -np.inf], np.nan).dropna()

    if len(long_inv_vol) == 0 or len(short_inv_vol) == 0:
        equity.append(equity[-1])
        continue

    # Normalize weights: (1/vol) / sum(same_side_1/vol) * 0.5
    long_weights = (long_inv_vol / long_inv_vol.sum()) * 0.5
    short_weights = (short_inv_vol / short_inv_vol.sum()) * -0.5  # Negative for short positions

    # Update current positions
    current_long_weights = long_weights
    current_short_weights = short_weights
    last_rebalance_date = current_date

    # Calculate weighted returns for the next day
    long_return = 0
    short_return = 0

    for coin in long_weights.index:
        p0 = price[coin].iloc[i]
        p1 = price[coin].iloc[i+1]
        if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
            coin_return = (p1 - p0) / p0
            long_return += long_weights[coin] * coin_return

    for coin in short_weights.index:
        p0 = price[coin].iloc[i]
        p1 = price[coin].iloc[i+1]
        if np.isfinite(p0) and p0 != 0 and np.isfinite(p1):
            coin_return = (p1 - p0) / p0
            short_return += short_weights[coin] * coin_return  # short_weights are already negative

    # Portfolio return: long + short - fees (short_return already negative)
    portfolio_return = long_return + short_return - 0.0005  # 5bps fees
    equity.append(equity[-1] * (1 + portfolio_return))
```

```python
plt.plot(price.index[:len(equity)], equity)
plt.xlabel("Date")
plt.ylabel("Equity")
plt.title("Strategy Equity Curve")
plt.xticks(rotation=45)
plt.show()
```
